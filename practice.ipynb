```python
import numpy as np

documents = [
    "I love machine learning",
    "I love coding in Python", 
    "Machine learning is fun"
]
def tokenize(text):
    text = text.lower()
    words = text.split()
    return words
def build_vocabulary(documents):
    vocabulary = set()
    for doc in documents:
        words = tokenize(doc)
        vocabulary.update(words)
    return list(vocabulary)
print(tokenize("I love Machine Learning"))
print(build_vocabulary(documents))

def create_term_frequency_matrix(documents, vocabulary):
    num_docs = len(documents)
    vocab_size = len(vocabulary)
    tf_matrix = np.zeros((num_docs, vocab_size))
    for i, doc in enumerate(documents):
        words = tokenize(doc)
        for word in words:
            if word in vocabulary:
                tf_matrix[i, vocabulary.index(word)] += 1
    return tf_matrix

vocab = build_vocabulary(documents)
tf_matrix = create_term_frequency_matrix(documents, vocab)
print("Vocabulary:", vocab)
print("TF Matrix shape:", tf_matrix.shape)
print("TF Matrix:\n", tf_matrix)

def calculate_idf(documents, vocabulary):
    num_docs = len(documents)
    idf_scores = np.zeros(len(vocabulary))
    for idx, word in enumerate(vocabulary):
        doc_count = sum(1 for doc in documents if word in tokenize(doc))
        idf_scores[idx] = np.log(num_docs / doc_count)
    return idf_scores

idf_scores = calculate_idf(documents, vocab)
print("IDF Scores:", idf_scores)
print("IDF shape:", idf_scores.shape)

def calculate_tfidf(tf_matrix, idf_scores):
    tfidf_matrix = np.zeros((tf_matrix.shape[0], tf_matrix.shape[1]))
    for i in range(tf_matrix.shape[0]):
        tfidf_matrix[i] = tf_matrix[i] * idf_scores
    return tfidf_matrix

tfidf_matrix = calculate_tfidf(tf_matrix, idf_scores)
print("TF-IDF Matrix shape:", tfidf_matrix.shape)
print("TF-IDF Matrix:\n", tfidf_matrix)

def cosine_similarity(a, b):
    return np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))

def find_most_similar_documents(tfidf_matrix, query_idx):
    query_vector = tfidf_matrix[query_idx]
    similarities = []
    for i in range(tfidf_matrix.shape[0]):
        if i != query_idx:
            similarity = cosine_similarity(query_vector, tfidf_matrix[i])
            similarities.append((i, similarity))
    return sorted(similarities, key=lambda x: x[1], reverse=True)

print("\n=== Document Similarity ===")
print("Documents:")
for i, doc in enumerate(documents):
    print(f"Doc {i}: {doc}")

print("\nFinding documents similar to Doc 0:")
similar = find_most_similar_documents(tfidf_matrix, 0)
print(similar)

print("\nFinding documents similar to Doc 1:")
similar = find_most_similar_documents(tfidf_matrix, 1)
print(similar)

print("\nFinding documents similar to Doc 2:")
similar = find_most_similar_documents(tfidf_matrix, 2)
print(similar)
    
```
